```python
import pandas as pd
import sklearn
```

```python
data = pd.read_csv("data/HR_comma_sep.csv")
data.head()
```

```python
data.columns
```

```python
# i have to examine the data according to the left column, that if the person has left the company or not
data.groupby("left").mean()
```

```python
# from above, i can see that only columns affecting the person based on the left attribute are satisfaction_level, promotion_last_5years, salary and 
# average_montly_hours
# so getting all the neccesary features for our model

df = data[["satisfaction_level", "promotion_last_5years", "average_montly_hours", "salary"]]
df.head()
```

```python
# there is one thing that we have to hot encode the salary feature as its value are in text format
dummy = pd.get_dummies(df.salary, prefix="salary")
x = pd.concat([df, dummy], axis=1)
x.head()
```

```python
# now removing the salary column as we have hot encoded the salary column
x= x.drop("salary", axis=1)
x.head()
```

```python
#getting the target data (that is left column) and splitting the data

y = data.left

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
